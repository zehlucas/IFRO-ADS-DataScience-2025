### Carregamento do dataset e bibliotecas

```python
import pandas as pd
dataset = pd.read_csv("../acidentes_2022.csv")
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
dataset_localidades = pd.read_csv("./localidades_2022.csv")
dataset_localidades.head()
```

```python
acidentes_RO = dataset[dataset['uf_acidente'] == "RO"]
```

```python
acidentes_RO.head()
```

```python
dataset['uf_acidente'] == "RO"
```

```python
acidentes_RO_Localidades = pd.merge(acidentes_RO, dataset_localidades, on="chv_localidade")
```

```python
acidentes_RO_Localidades.head()
```

```python
acidentes_RO_Localidades.groupby('municipio').count()['num_acidente'].sort_values(ascending=False)
```
